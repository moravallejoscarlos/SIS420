```python
# datos = estado
class Nodo:
    def __init__(self, datos, hijo=None):
        self.datos = datos
        self.hijos = []
        self.padre = None
        self.costo = None
        self.set_hijo(hijo)
        
    def set_hijo(self, hijo):
        if (hijo is not None):
            self.hijos.append(hijo)
            if self.hijos is not None:
                for h in self.hijos:
                    h.padre = self
                
    def get_hijos(self):
        return self.hijos
    
    def set_padre(self, padre):
        self.padre = padre
        
    def get_padre(self):
        return self.padre

    def set_datos(self, datos):
        self.datos = datos
    
    def get_datos(self):
        return self.datos
    
    def set_costo(self, costo):
        self.costo = costo
        
    def get_costo(self):
        return self.costo
    
    def equal(self, nodo):
        if self.get_datos() == nodo.get_datos():
            return True
        else:
            return False
    
    def en_lista(self, lista_nodos):
        enlistado = False
        for n in lista_nodos:
            if self.equal(n):
                enlistado = True
        return enlistado
    
    def __str__(self):
        return str(self.get_datos())
```

```python
def bpa(estado_inicio, estado_solucion):
    resuelto = False
    nodos_visitados = []
    nodos_frontera = []
    nodo_inicio = Nodo(estado_inicio)
    nodos_frontera.append(nodo_inicio)
    
    while resuelto == False and len(nodos_frontera) != 0:
        nodo_actual = nodos_frontera.pop(0)
        nodos_visitados.append(nodo_actual)
        if nodo_actual.get_datos() == estado_solucion:
            resuelto = True
            return nodo_actual
        else:
            hijo1_datos = nodo_actual.get_datos().copy()
            temp = hijo1_datos[0]
            hijo1_datos[0] = hijo1_datos[1]
            hijo1_datos[1] = temp
            hijo1 = Nodo(hijo1_datos)
            
            if not hijo1.en_lista(nodos_visitados) and not hijo1.en_lista(nodos_frontera):
                nodo_actual.set_hijo(hijo1)
                nodos_frontera.append(hijo1)
                
            hijo2_datos = nodo_actual.get_datos().copy()
            temp = hijo2_datos[1]
            hijo2_datos[1] = hijo2_datos[2]
            hijo2_datos[2] = temp
            hijo2 = Nodo(hijo2_datos)
            if not hijo2.en_lista(nodos_visitados) and not hijo2.en_lista(nodos_frontera):
                nodos_frontera.append(hijo2)
                nodo_actual.set_hijo(hijo2)
                
            hijo3_datos = nodo_actual.get_datos().copy()
            temp = hijo3_datos[2]
            hijo3_datos[2] = hijo3_datos[3]
            hijo3_datos[3] = temp
            hijo3 = Nodo(hijo3_datos)
            if not hijo3.en_lista(nodos_visitados) and not hijo3.en_lista(nodos_frontera):
                nodos_frontera.append(hijo3)
                nodo_actual.set_hijo(hijo3) 

            hijo4_datos = nodo_actual.get_datos().copy()
            temp = hijo4_datos[3]
            hijo4_datos[3] = hijo4_datos[4]
            hijo4_datos[4] = temp
            hijo4 = Nodo(hijo4_datos)
            if not hijo4.en_lista(nodos_visitados) and not hijo4.en_lista(nodos_frontera):
                nodos_frontera.append(hijo4)
                nodo_actual.set_hijo(hijo4) 

            hijo5_datos = nodo_actual.get_datos().copy()
            temp = hijo5_datos[4]
            hijo5_datos[4] = hijo5_datos[5]
            hijo5_datos[5] = temp
            hijo5 = Nodo(hijo5_datos)
            if not hijo5.en_lista(nodos_visitados) and not hijo5.en_lista(nodos_frontera):
                nodos_frontera.append(hijo5)
                nodo_actual.set_hijo(hijo5)     

            hijo6_datos = nodo_actual.get_datos().copy()
            temp = hijo6_datos[5]
            hijo6_datos[5] = hijo6_datos[6]
            hijo6_datos[6] = temp
            hijo6 = Nodo(hijo6_datos)
            if not hijo6.en_lista(nodos_visitados) and not hijo6.en_lista(nodos_frontera):
                nodos_frontera.append(hijo6)
                nodo_actual.set_hijo(hijo6)     

            hijo7_datos = nodo_actual.get_datos().copy()
            temp = hijo3_datos[6]
            hijo7_datos[6] = hijo7_datos[7]
            hijo7_datos[7] = temp
            hijo7 = Nodo(hijo7_datos)
            if not hijo7.en_lista(nodos_visitados) and not hijo7.en_lista(nodos_frontera):
                nodos_frontera.append(hijo7)
                nodo_actual.set_hijo(hijo7)     

            hijo8_datos = nodo_actual.get_datos().copy()
            temp = hijo3_datos[7]
            hijo8_datos[7] = hijo8_datos[8]
            hijo8_datos[8] = temp
            hijo8 = Nodo(hijo8_datos)
            if not hijo8.en_lista(nodos_visitados) and not hijo8.en_lista(nodos_frontera):
                nodos_frontera.append(hijo8)
                nodo_actual.set_hijo(hijo8)     

            hijo9_datos = nodo_actual.get_datos().copy()
            temp = hijo3_datos[8]
            hijo9_datos[8] = hijo9_datos[9]
            hijo9_datos[9] = temp
            hijo9 = Nodo(hijo9_datos)
            if not hijo9.en_lista(nodos_visitados) and not hijo9.en_lista(nodos_frontera):
                nodos_frontera.append(hijo9)
                nodo_actual.set_hijo(hijo9)     

            hijo10_datos = nodo_actual.get_datos().copy()
            temp = hijo3_datos[9]
            hijo10_datos[9] = hijo10_datos[10]
            hijo10_datos[10] = temp
            hijo10 = Nodo(hijo10_datos)
            if not hijo10.en_lista(nodos_visitados) and not hijo10.en_lista(nodos_frontera):
                nodos_frontera.append(hijo10)
                nodo_actual.set_hijo(hijo10)

             # la maquina se colapsa que no se puede superar mayor que 7 y tardaria mucho tiempo aunque tenga buenos componetes en tu computadora  
```
