```python
class Nodo:
    def __init__(self, estado, hijo=None):
        self.estado = estado
        self.hijo = None
        self.padre = None
        self.accion = None
        self.acciones = None
        self.costo = None
        self.set_hijo(hijo)

    def set_estado(self, estado):
        self.estado = estado

    def get_estado(self):
        return self.estado

    def set_hijo(self, hijo):
        self.hijo = hijo
        if self.hijo is not None:
            for s in self.hijo:
                s.padre = self

    def get_hijo(self):
        return self.hijo
    
    def set_padre(self, padre):
        self.padre = padre

    def get_padre(self):
        return self.padre
    
    def set_accion(self, accion):
        self.padre = accion

    def get_accion(self):
        return self.accion

    def set_acciones(self, acciones):
        self.acciones = acciones

    def get_acciones(self):
        return self.acciones

    def set_costo(self, costo):
        self.costo = costo

    def get_costo(self):
        return self.costo

    def equal(self, Nodo):
        if self.get_estado() == Nodo.get_estado():
            return True
        else:
            return False

    def en_lista(self, lista_nodos):
        enlistado = False
        for n in lista_nodos:
            if self.equal(n):
                enlistado = True
        return enlistado

    def __str__(self):
        return str(self.get_estado())
```

```python
"""
# movimientos posibles
torre1 a torre2
torre1 a torre3
torre2 a torre1
torre2 a torre3
torre3 a torre1
torre3 a torre2
"""
torre = [[0,0,3],[0,2,1],[0,0,0]];
def valid_mov(torre):
    invalid=0
    valid =0
    for i in range(3):
        # verificar si esta ordenado de menor a mayor contado 0 como el menor
        if torre[i][0] <= torre[i][1] <= torre[i][2]:
            #print ("torre " + str(i) + " ordenada")
            valid +=1
        else:
            #print("torre " + str(i) + " desordenada")
            invalid=1
    if(invalid>0):
        return False
    else:
        return True
```

```python
def buscar_primer_numero(torre):
    for i, v in enumerate(torre):
        
        if(v!=0):
            return v, i
#buscar 0 antes del primer numero
def buscar_0(torre):
    if (torre==[0,0,0]):
        return 2
    else:
        for i, v in enumerate(torre):
            if(v>0):
                return i-1
```

```python
#generar hijos de toore 1 a torre 2
def move_t1_t2(torre):
    hijo1 = torre.copy()
    valor=0
    idx = 0
    if (torre[0]==[0,0,0]):
        return hijo1
    valor, idx = buscar_primer_numero(torre[0])
    hijo1[0][idx]=0
    idx = buscar_0(torre[1])
    hijo1[1][idx]=valor
    return(hijo1)

#generar hijos de toore 1 a torre 3
def move_t1_t3(torre):
    hijo1 = torre.copy()
    valor=0
    idx = 0
    if (torre[0]==[0,0,0]):
        return hijo1
    valor, idx = buscar_primer_numero(torre[0])
    hijo1[0][idx]=0
    idx = buscar_0(torre[2])
    hijo1[2][idx]=valor
    return(hijo1)

#generar hijos de toore 2 a torre 1
def move_t2_t1(torre):
    hijo1 = torre.copy()
    valor=0
    idx = 0
    if (torre[1]==[0,0,0]):
        return hijo1
    valor, idx = buscar_primer_numero(torre[1])
    hijo1[1][idx]=0
    idx = buscar_0(torre[0])
    hijo1[0][idx]=valor
    return(hijo1)

#generar hijos de toore 2 a torre 3
def move_t2_t3(torre):
    hijo1 = torre.copy()
    idx = 0
    valor=0
    if (torre[1]==[0,0,0]):
        return hijo1
    valor, idx = buscar_primer_numero(torre[1])
   
    hijo1[1][idx]=0
    idx = buscar_0(torre[2])
    hijo1[2][idx]=valor
    return(hijo1)

#generar hijos de toore 3 a torre 1
def move_t3_t1(torre):
    hijo1 = torre.copy()
    valor=0
    idx = 0
    if (torre[2]==[0,0,0]):
        return hijo1
    valor, idx = buscar_primer_numero(torre[2])
    hijo1[2][idx]=0
    idx = buscar_0(torre[0])
    hijo1[0][idx]=valor
    return(hijo1)

#generar hijos de toore 3 a torre 2
def move_t3_t2(torre):
    hijo1 = torre.copy()
    valor=0
    idx = 0
    if (torre[2]==[0,0,0]):
        return hijo1
    valor, idx = buscar_primer_numero(torre[2])
    hijo1[2][idx]=0
    idx = buscar_0(torre[1])
    hijo1[1][idx]=valor
    return(hijo1)
```

```python
def busqueda_BPA_solucion(estado_inicial, solucion):
    resuelto = False
    nodos_visitados = []
    nodos_frontera = []

    nodo_raiz = Nodo(estado_inicial)
    nodos_frontera.append(nodo_raiz)
    while (not resuelto) and len(nodos_frontera) != 0:
        nodo_actual = nodos_frontera.pop(0)
        # extraer nodo y a√±adirlo a visitados
        nodos_visitados.append(nodo_actual)
        if nodo_actual.get_estado() == solucion:
            # solucion encontrada
            resuelto = True
            return nodo_actual
        else:
            # expandir nodos hijo
            estado_nodo = nodo_actual.get_estado()
            hijos =[]

            # movimientos posibles
            """
            torre1 a torre2
            torre1 a torre3
            torre2 a torre1
            torre2 a torre3
            torre3 a torre1
            torre3 a torre2
            """

            # torre1 a torre2
            #hijo = [[0,0,0],[0,0,0],[0,0,0]]
            hijo = move_t1_t2([[estado_nodo[0][0],estado_nodo[0][1],estado_nodo[0][2]],[estado_nodo[1][0],estado_nodo[1][1],estado_nodo[1][2]],[estado_nodo[2][0],estado_nodo[2][1],estado_nodo[2][2]]])
            
            hijo_1 = Nodo(hijo)
            if(valid_mov(hijo) and hijo != estado_nodo):
                hijos.append(hijo_1)
            if not hijo_1.en_lista(nodos_visitados) and not hijo_1.en_lista(nodos_frontera):
                nodos_frontera.append(hijo_1)
            
            # torre1 a torre3
            hijo = move_t1_t3([[estado_nodo[0][0],estado_nodo[0][1],estado_nodo[0][2]],[estado_nodo[1][0],estado_nodo[1][1],estado_nodo[1][2]],[estado_nodo[2][0],estado_nodo[2][1],estado_nodo[2][2]]])
            hijo_2 = Nodo(hijo)
            if(valid_mov(hijo) and hijo != estado_nodo):
                hijos.append(hijo_2)
            if not hijo_2.en_lista(nodos_visitados) and not hijo_2.en_lista(nodos_frontera):
                nodos_frontera.append(hijo_2)

            # torre2 a torre1
            hijo = move_t2_t1([[estado_nodo[0][0],estado_nodo[0][1],estado_nodo[0][2]],[estado_nodo[1][0],estado_nodo[1][1],estado_nodo[1][2]],[estado_nodo[2][0],estado_nodo[2][1],estado_nodo[2][2]]])
            hijo_3 = Nodo(hijo)
            if(valid_mov(hijo) and hijo != estado_nodo):
                hijos.append(hijo_3)
            if not hijo_3.en_lista(nodos_visitados) and not hijo_3.en_lista(nodos_frontera):
                nodos_frontera.append(hijo_3)

            # torre2 a torre3
            hijo = move_t2_t3([[estado_nodo[0][0],estado_nodo[0][1],estado_nodo[0][2]],[estado_nodo[1][0],estado_nodo[1][1],estado_nodo[1][2]],[estado_nodo[2][0],estado_nodo[2][1],estado_nodo[2][2]]])
            hijo_4 = Nodo(hijo)
            if(valid_mov(hijo) and hijo != estado_nodo):
                hijos.append(hijo_4)
            if not hijo_4.en_lista(nodos_visitados) and not hijo_4.en_lista(nodos_frontera):
                nodos_frontera.append(hijo_4)

            # torre3 a torre1
            hijo = move_t3_t1([[estado_nodo[0][0],estado_nodo[0][1],estado_nodo[0][2]],[estado_nodo[1][0],estado_nodo[1][1],estado_nodo[1][2]],[estado_nodo[2][0],estado_nodo[2][1],estado_nodo[2][2]]])
            hijo_5 = Nodo(hijo)
            if(valid_mov(hijo) and hijo != estado_nodo):
                hijos.append(hijo_5)
            if not hijo_5.en_lista(nodos_visitados) and not hijo_5.en_lista(nodos_frontera):
                nodos_frontera.append(hijo_5)

            # torre3 a torre2
            hijo = move_t3_t2([[estado_nodo[0][0],estado_nodo[0][1],estado_nodo[0][2]],[estado_nodo[1][0],estado_nodo[1][1],estado_nodo[1][2]],[estado_nodo[2][0],estado_nodo[2][1],estado_nodo[2][2]]])
            hijo_6 = Nodo(hijo)
            if(valid_mov(hijo) and hijo != estado_nodo):
                hijos.append(hijo_6)
            if not hijo_6.en_lista(nodos_visitados) and not hijo_6.en_lista(nodos_frontera):
                nodos_frontera.append(hijo_6)

            
            nodo_actual.set_hijo(hijos)
```
